```python
plt.plot([1, 4, 9, 16])
# 1. Matplotlib의 pyplot 모듈(plt)을 사용하여 `plot` 함수를 호출합니다.
# 2. 인수로 하나의 리스트 `[1, 4, 9, 16]`만 제공되었습니다.
# 3. Matplotlib은 이 값을 "y축 데이터"로 간주합니다.
# 4. x축 데이터는 자동으로 "0부터 시작하는 정수 인덱스"로 생성됩니다.
#    - (1의 x좌표는 0, 4의 x좌표는 1, 9의 x좌표는 2, 16의 x좌표는 3)
# 5. 이 점들을 이어서 꺾은선 그래프(Line Plot)를 준비합니다. 

plt.show()
# 6. 준비된 그래프를 화면에 "출력"합니다. 이 함수를 호출해야 실제로 창이 나타나 그래프를 볼 수 있습니다.
```

```python
t = [0, 1, 2, 3, 4, 5, 6] # x축 데이터(독립 변수) 리스트를 정의합니다. (시간, time 등)
y = [1, 4, 5, 8, 9, 5, 3] # y축 데이터(종속 변수) 리스트를 정의합니다. (값, value 등)

plt.figure(figsize=(10, 6))
# 1. 새로운 그래프 창(Figure)을 생성합니다.
# 2. `figsize=(10,6)` 옵션: 그래프 창의 크기를 "가로 10인치, 세로 6인치"로 설정합니다.
# (그래프를 더 크고 보기 좋게 만듭니다.)

plt.plot(t, y, color='green')
# 3. `plot` 함수를 호출하여 꺾은선 그래프를 그립니다.
# 4. 첫 번째 인수로 `t` (x축 데이터), 두 번째 인수로 `y` (y축 데이터)를 명시적으로 지정합니다.
# 5. `color='green'` 옵션: 그려질 선의 색상을 "녹색"으로 지정합니다.

plt.show()
# 6. 생성된 그래프를 화면에 출력합니다.
```

```python
t = np.arange(0., 5., 0.2)
# NumPy의 arange 함수를 사용하여 x축 데이터 배열 t를 생성합니다.
# t는 0.0부터 5.0 미만까지 0.2 간격으로 증가하는 값을 갖습니다.
# (예: [0.0, 0.2, 0.4, ..., 4.8])

print(t) # 생성된 배열 t의 내용을 출력합니다.

plt.figure(figsize=(10,6)) # 새로운 그림(figure) 객체를 생성하고, 그래프 크기를 가로 10인치, 세로 6인치로 설정합니다.
plt.title('라인플롯에서 여러개의 선 그리기') # 그래프의 제목을 설정합니다.

plt.plot(t, t, 'r--')
# 첫 번째 선을 그립니다. (y = t, 즉 일차 함수)
# 'r--': 빨간색(red)의 대시선(dashed line) 스타일로 지정합니다.

plt.plot(t, 0.5*t**2, 'bs:')
# 두 번째 선을 그립니다. (y = 0.5t^2, 즉 이차 함수)
# 'bs:': 파란색(blue), 사각형(square) 마커, 그리고 점선(dotted line) 스타일로 지정합니다.

plt.plot(t, 0.2*t**3, 'g^-')
# 세 번째 선을 그립니다. (y = 0.2t^3, 즉 삼차 함수)
# 'g^-': 녹색(green), 위쪽 삼각형(triangle up) 마커, 그리고 실선(solid line) 스타일로 지정합니다.

plt.show() # 생성된 그래프를 화면에 표시합니다.
```

```python
t = np.arange(0., 5., 0.2)
# NumPy의 arange 함수를 사용하여 x축 데이터 배열 t를 생성합니다.
# t는 0.0부터 5.0 미만까지 0.2 간격으로 증가하는 값을 갖습니다.
# (예: [0.0, 0.2, 0.4, ..., 4.8])

plt.plot(t, t, 'r--',         # 첫 번째 (x, y, format) 쌍:
                               # x축 데이터: t, y축 데이터: t (y=t, 일차 함수)
                               # 'r--': 빨간색(red)의 대시선(dashed line) 스타일로 지정합니다.
         t, 0.5*t**2, 'bs:',  # 두 번째 (x, y, format) 쌍:
                               # x축 데이터: t, y축 데이터: 0.5*t**2 (y=0.5t^2, 이차 함수)
                               # 'bs:': 파란색(blue), 사각형(square) 마커, 그리고 점선(dotted line) 스타일로 지정합니다.
         t, 0.2*t**3, 'g^-')  # 세 번째 (x, y, format) 쌍:
                               # x축 데이터: t, y축 데이터: 0.2*t**3 (y=0.2t^3, 삼차 함수)
                               # 'g^-': 녹색(green), 위쪽 삼각형(triangle up) 마커, 그리고 실선(solid line) 스타일로 지정합니다.

plt.show() # 생성된 그래프를 화면에 표시합니다.
```

```python
df1 = pd.DataFrame({
    '나이' : [15, 20, 17, 50, 2, 30, 23], # '나이' 열의 데이터 리스트
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희'] # '이름' 열의 데이터 리스트
}, columns=['나이', '이름']) # '나이'와 '이름' 두 개의 열을 가진 DataFrame을 생성합니다.

# --- 첫 번째 .plot() 호출 (막대 그래프) ---
df1.plot('이름', '나이', kind='bar')
# DataFrame의 .plot() 메서드를 사용하여 **막대 그래프**를 그립니다.
# 첫 번째 인수 ('이름'): X축 데이터로 사용될 열을 지정합니다. (레이블로 사용됨)
# 두 번째 인수 ('나이'): Y축 데이터로 사용될 열을 지정합니다. (막대 높이)
# kind='bar': 막대 그래프 유형을 지정합니다.
# 이 호출로 인해 첫 번째 Figure가 생성됩니다.

# --- 두 번째 .plot() 호출 (선 그래프) ---
df1.plot(y='나이').plot(grid=True, figsize=(5, 5))
# 이 줄은 복잡하게 보이지만, 결국 '나이' 열에 대한 두 번째 Figure를 생성합니다.
# 1. df1.plot(y='나이'): '나이' 열에 대한 **기본 선 그래프**를 그립니다. (plot의 기본값은 'line')
# 2. .plot(grid=True, figsize=(5,5)): 첫 번째 .plot() 호출이 Axes 객체를 반환하며,
#    이 Axes 객체에는 .plot() 메서드가 없습니다. 따라서 이 부분은 **오류**가 발생하거나,
#    혹은 Matplotlib의 내부 상태에 따라 **예상치 못한 결과를 초래**할 수 있습니다.
#    (일반적으로 Pandas의 연속적인 .plot() 호출은 지원되지 않음).
#    *의도된 해석:* 새로운 Figure에 '나이' 열에 대한 **선 그래프**를 그리려 했으며, 격자와 크기를 설정하려 했습니다.
#    올바른 문법: `df1.plot(y='나이', kind='line', grid=True, figsize=(5,5))`

plt.show() # 생성된 모든 Figure를 화면에 표시합니다.
# 만약 두 번째 .plot() 호출에서 오류가 발생하지 않았다면, 두 개의 독립적인 그래프 창이 나타날 것입니다.
```

```python
N=30                 # 데이터 포인트의 개수(N)를 30으로 설정합니다.
np.random.seed(0)    # 난수 생성기 시드(seed)를 0으로 고정하여 결과 재현성을 보장합니다.

x = np.random.rand(N) # 0과 1 사이의 난수 30개로 이루어진 x축 좌표 배열을 생성합니다. (위치 결정)
y1 = np.random.rand(N) # 0과 1 사이의 난수 30개로 이루어진 y축 좌표 배열을 생성합니다. (위치 결정)
y2 = np.random.rand(N) # 0과 1 사이의 난수 30개로 이루어진 배열을 생성합니다. (색상 결정)

y3 = np.pi * (15 * np.random.rand(N)) ** 2
# 마커의 크기(area)를 결정하는 배열을 생성합니다. (크기 결정)
# 1. np.random.rand(N): 0~1 사이의 난수 30개 생성.
# 2. 15 * (...): 값을 0~15 사이로 확장합니다.
# 3. (...) ** 2: 제곱하여 마커의 지름이 아닌 면적(Area)에 비례하도록 만듭니다.
# 4. np.pi * (...): 최종적으로 면적($\pi r^2$)과 관련된 형태로 조정됩니다.

# --- 산점도 (버블 차트) 생성 ---
plt.scatter(x, y1, c=y2, s=y3)
# plt.scatter()를 사용하여 산점도/버블 차트를 그립니다.
# x: x축 위치 (x 배열)
# y1: y축 위치 (y1 배열)
# c=y2: 마커의 색상(color)을 y2 배열의 값에 따라 다르게 설정합니다. (Colormap에 매핑됨)
# s=y3: 마커의 크기(size)를 y3 배열의 값에 따라 다르게 설정합니다. (면적에 해당)

plt.show() # 생성된 버블 차트 그래프를 화면에 표시합니다.
```

```python
np.random.seed(0) # NumPy의 난수 생성기 시드(seed)를 0으로 고정합니다. (결과 재현성을 보장)

x = np.random.randn(1000)
# 표준 정규 분포(평균 0, 표준 편차 1)를 따르는 난수 1000개를 생성하여 x 배열에 저장합니다.
# 이 데이터는 종 모양의 분포(Bell Curve)를 가집니다.

plt.title('히스토그램') # 그래프의 제목을 '히스토그램'으로 설정합니다.

plt.hist(x)
# 히스토그램을 생성하고 그립니다.
# x: 분석할 데이터 배열입니다.
# Matplotlib은 x 데이터의 최소값부터 최대값까지의 범위를 "10개의 구간(bins)"으로 자동 분할하고,
# 각 구간에 속하는 데이터의 개수(빈도)를 막대의 높이로 표시합니다.

plt.show() # 생성된 히스토그램 그래프를 화면에 표시합니다.
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
# 's1' 데이터 배열을 생성합니다.
# loc=0: 분포의 평균(mean)을 0으로 설정합니다.
# scale=1: 분포의 표준 편차(standard deviation, std)를 1로 설정합니다.
# size=1000: 1000개의 데이터 포인트를 생성합니다.

s2 = np.random.normal(loc=5, scale=0.5, size=1000)
# 's2' 데이터 배열을 생성합니다.
# loc=5: 분포의 평균을 5로 설정합니다.
# scale=0.5: 분포의 표준 편차를 0.5로 설정합니다. (s1보다 더 좁게 분포)
# size=1000: 1000개의 데이터 포인트를 생성합니다.

s3 = np.random.normal(loc=10, scale=2, size=1000)
# 's3' 데이터 배열을 생성합니다.
# loc=10: 분포의 평균을 10으로 설정합니다.
# scale=2: 분포의 표준 편차를 2로 설정합니다. (s1, s2보다 가장 넓게 분포)
# size=1000: 1000개의 데이터 포인트를 생성합니다.


plt.plot(s1, label='s1')
# s1 배열의 값을 Y축으로 하는 선 그래프를 그립니다. (X축은 0부터 999까지의 인덱스)
# label='s1': 범례에 표시될 레이블을 설정합니다.

plt.plot(s2, label='s2')
# s2 배열의 값을 Y축으로 하는 선 그래프를 그립니다.

plt.plot(s3, label='s3')
# s3 배열의 값을 Y축으로 하는 선 그래프를 그립니다.

plt.legend()
# 각 plt.plot() 함수에서 설정된 label을 바탕으로 범례(Legend)를 그래프에 표시합니다.

plt.show() # 생성된 그래프를 화면에 표시합니다.
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
# 's1': 평균(loc) 0, 표준 편차(scale) 1인 데이터 1000개 생성.

s2 = np.random.normal(loc=5, scale=0.5, size=1000)
# 's2': 평균(loc) 5, 표준 편차(scale) 0.5인 데이터 1000개 생성. (s1보다 밀집)

s3 = np.random.normal(loc=10, scale=2, size=1000)
# 's3': 평균(loc) 10, 표준 편차(scale) 2인 데이터 1000개 생성. (s1, s2보다 분산 큼)

plt.figure(figsize=(10,6)) # 새로운 그림(Figure) 객체를 생성하고, 크기를 설정합니다.

plt.boxplot([s1,s2,s3])
# 박스 플롯을 그립니다. 인수로 여러 데이터 배열([s1, s2, s3])을 리스트 형태로 전달하여,
# 하나의 그래프에 여러 데이터셋의 분포를 나란히 비교합니다.

plt.grid() # 그래프 배경에 격자(Grid)를 추가하여 값의 가독성을 높입니다.

plt.show() # 생성된 박스 플롯 그래프를 화면에 표시합니다.
```

```python
labels = ['개구리', '돼지', '개', '통나무'] # 각 조각(slice)에 표시될 이름(레이블) 목록
size = [15, 30, 45, 10]                 # 각 조각의 크기(값) 목록 (총 합은 100)
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'] # 각 조각에 적용될 색상 목록

plt.figure(figsize=(10, 6)) # 새로운 그림(Figure) 객체를 생성하고, 크기를 설정합니다.

plt.title('Pie Chart') # 그래프의 제목을 'Pie Chart'로 설정합니다.

plt.pie(size, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
# 파이 차트를 그립니다.
# size: 각 조각의 크기(비율)를 결정하는 데이터입니다. (15 + 30 + 45 + 10 = 100)
# labels: 각 조각에 대한 레이블(이름)을 설정합니다.
# colors: 각 조각에 대한 색상을 설정합니다.
# autopct='%1.1f%%': 각 조각 내부에 "자동으로 비율(%)"을 표시합니다.
#                   '%1.1f%%'는 소수점 이하 첫째 자리까지 표시하도록 형식을 지정합니다.
# shadow=True: 파이 차트 아래에 그림자를 추가하여 입체감을 부여합니다.
# startangle=90: 첫 번째 조각('개구리')이 "90도 (12시 방향)"에서 시작하도록 회전 각도를 설정합니다.

plt.show() # 생성된 파이 차트 그래프를 화면에 표시합니다.
```
