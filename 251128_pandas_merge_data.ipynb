```python
# --- 6. Left/Right Join (ê¸°ì¤€ í…Œì´ë¸” ë³‘í•©) ---

# how='left'ëŠ” "df1 (ì™¼ìª½ í…Œì´ë¸”)"ì˜ ëª¨ë“  'ê³ ê°ë²ˆí˜¸'ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©í•©ë‹ˆë‹¤.
# (df1ì˜ ëª¨ë“  ê³ ê°(1001~1007)ì€ í¬í•¨ë˜ë©°, ì§ì´ ì—†ëŠ” df2ì˜ 1008ì€ ì œì™¸ë©ë‹ˆë‹¤.)
# (df2ì— ì§ì´ ì—†ëŠ” df1ì˜ í–‰ì€ ê¸ˆì•¡ ì—´ì´ NaNìœ¼ë¡œ ì±„ì›Œì§‘ë‹ˆë‹¤.)
print(pd.merge(df1, df2, how='left')) 
```

```python
# how='right'ëŠ” "df2 (ì˜¤ë¥¸ìª½ í…Œì´ë¸”)"ì˜ ëª¨ë“  'ê³ ê°ë²ˆí˜¸'ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©í•©ë‹ˆë‹¤.
# (df2ì— ìˆëŠ” ëª¨ë“  ê³ ê°(1001, 1005, 1006, 1008)ì€ í¬í•¨ë˜ë©°, ì§ì´ ì—†ëŠ” df1ì˜ ê³ ê°ì€ ì œì™¸ë©ë‹ˆë‹¤.)
# (df1ì— ì§ì´ ì—†ëŠ” df2ì˜ í–‰(1008)ì€ ì´ë¦„ ì—´ì´ NaNìœ¼ë¡œ ì±„ì›Œì§‘ë‹ˆë‹¤.)
print(pd.merge(df1, df2, how='right')) 
# ê²°ê³¼: "ì˜¤ë¥¸ìª½ DataFrame (df2)ì˜ ëª¨ë“  í–‰"ì„ ìœ ì§€í•˜ê³ , ì™¼ìª½ df1ì˜ ì¼ì¹˜í•˜ëŠ” í–‰ì„ ê²°í•©í•©ë‹ˆë‹¤.
# - df2ì˜ 6ê°œ êµ¬ë§¤ ë‚´ì—­ì€ ëª¨ë‘ ë³´ì¡´ë©ë‹ˆë‹¤.
# - df1ì—ë§Œ ìˆëŠ” ê³ ê°(1002, 1003, 1004, 1007)ì€ ì œì™¸ë©ë‹ˆë‹¤.
# - df2ì—ë§Œ ìˆëŠ” ê³ ê° 1008ì˜ 'ì´ë¦„'ì€ NaN(ê²°ì¸¡ì¹˜)ìœ¼ë¡œ ì±„ì›Œì§‘ë‹ˆë‹¤.
```

```python
# --- ì²« ë²ˆì§¸ DataFrame (df1) ìƒì„± ---

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']},
    index=[0, 1, 2, 3]) # ì—´: A, B, C, D / ì¸ë±ìŠ¤: 0, 1, 2, 3
print(df1)
```

```python
# --- ë‘ ë²ˆì§¸ DataFrame (df2) ìƒì„± ---

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'E': ['C4', 'C5', 'C6', 'C7'], # Cì™€ D ëŒ€ì‹  Eì™€ F ì—´ì„ ê°€ì§
    'F': ['D4', 'D5', 'D6', 'D7']},
    index=[0, 1, 2, 3]) # ì—´: A, B, E, F / ì¸ë±ìŠ¤: 0, 1, 2, 3
print(df2)
```

```python
# --- ì„¸ ë²ˆì§¸ DataFrame (df3) ìƒì„± ---

df3 = pd.DataFrame({
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'O': ['D8', 'D9', 'D10', 'D11']}, # D ëŒ€ì‹  O ì—´ì„ ê°€ì§
    index=[1,2,3,4]) # ì—´: A, B, C, O / ì¸ë±ìŠ¤: 1, 2, 3, 4
print(df3)
```

```python
# # 1. ê¸°ë³¸ ì—°ê²° (Outer Join, ì¸ë±ìŠ¤ ì¤‘ë³µ)

result = pd.concat([df1, df2, df3]) # ê¸°ë³¸ê°’: axis=0 (ìˆ˜ì§), join='outer' (ì „ì²´ ì—´ ë³´ì¡´), ignore_index=False
print(result)
# ê²°ê³¼: ì„¸ DataFrameì˜ ëª¨ë“  í–‰ì´ ìˆœì„œëŒ€ë¡œ ì—°ê²°ë©ë‹ˆë‹¤.
#       - ì—´: ëª¨ë“  DataFrameì— ì¡´ì¬í•˜ëŠ” ëª¨ë“  ì—´(A, B, C, D, E, F, O)ì´ ë³´ì¡´ë©ë‹ˆë‹¤.
#       - ì—´ì´ ì—†ëŠ” ë¶€ë¶„ì€ NaNìœ¼ë¡œ ì±„ì›Œì§‘ë‹ˆë‹¤.
#       - ì¸ë±ìŠ¤: ì›ë³¸ ì¸ë±ìŠ¤(0, 1, 2, 3, 0, 1, 2, 3, 1, 2, 3, 4)ê°€ ê·¸ëŒ€ë¡œ ìœ ì§€ë˜ì–´ "ì¸ë±ìŠ¤ê°€ ì¤‘ë³µ"ë©ë‹ˆë‹¤.
```

```python
# # 2. MultiIndex ìƒì„± (keys ì‚¬ìš©)

result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
# ì—°ê²°ëœ ê° DataFrameì— ìµœìƒìœ„ ì¸ë±ìŠ¤('x', 'y', 'z')ë¥¼ ë¶€ì—¬í•©ë‹ˆë‹¤.
# ê²°ê³¼ëŠ” 2 ë ˆë²¨ì˜ MultiIndexë¥¼ ê°–ê²Œ ë©ë‹ˆë‹¤. (ìµœìƒìœ„: x, y, z / í•˜ìœ„: ì›ë³¸ ì¸ë±ìŠ¤)

print(result.loc['x'].loc[0, 'A'])
# MultiIndexì—ì„œ 'x' ê·¸ë£¹ì„ ì„ íƒí•˜ê³ , ê·¸ ì•ˆì—ì„œ ì¸ë±ìŠ¤ 0ì˜ 'A' ì—´ ê°’ì„ ì¶œë ¥í•©ë‹ˆë‹¤. ('A0')
```

```python
# # 3. Inner Join ì—°ê²° (join='inner')

result = pd.concat([df1, df2, df3], join='inner', keys=['x', 'y', 'z'])
print(result)
# ê²°ê³¼: ì„¸ DataFrameì— "ëª¨ë‘ ì¡´ì¬í•˜ëŠ” ê³µí†µ ì—´"ë§Œì„ ë‚¨ê¸°ê³  ì—°ê²°í•©ë‹ˆë‹¤.
# - df1(A, B, C, D), df2(A, B, E, F), df3(A, B, C, O)ì˜ ê³µí†µ ì—´ì€ "'A'ì™€ 'B'"ë¿ì…ë‹ˆë‹¤.
# - 'A'ì™€ 'B' ì—´ë§Œ ë‚¨ê³ , MultiIndex('x', 'y', 'z')ëŠ” ìœ ì§€ë©ë‹ˆë‹¤.
```

```python
# # 4. Inner Join ì—°ê²° (keys ì—†ìŒ)

result = pd.concat([df1, df2, df3], join='inner')
print(result)
# ê²°ê³¼: (3ë²ˆê³¼ ë™ì¼í•˜ê²Œ) ê³µí†µ ì—´('A', 'B')ë§Œ ë‚¨ê³  ì—°ê²°ë©ë‹ˆë‹¤.
# - MultiIndexëŠ” ìƒì„±ë˜ì§€ ì•Šê³ , ì›ë³¸ì˜ ì¤‘ë³µ ì¸ë±ìŠ¤(0, 1, 2, 3, 0, 1, 2, 3, 1, 2, 3, 4)ê°€ ê·¸ëŒ€ë¡œ ìœ ì§€ë©ë‹ˆë‹¤.
```
