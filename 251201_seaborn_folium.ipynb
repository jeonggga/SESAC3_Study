```python
# sns.despine() 함수
# 그래프의 '틀(테두리)'처럼 보이는 선들을 보이게 할지 말지 정하는 함수
# 예)
#   right=True   → 오른쪽 선을 숨김
#   right=False  → 오른쪽 선을 보이게 함
#   left=True    → 왼쪽 선을 숨김
#   top=True     → 위쪽 선을 숨김
#   bottom=False → 아래쪽 선을 보이게 함
# True = 숨김 / False = 보임

# x 값 만들기 (1부터 14까지 100개 숫자)
x = np.linspace(1, 14, 100)

# x를 이용해서 파도 모양 그래프 만들기 (sin 함수 이용)
y1 = np.sin(x)        # 1배 크기 파도
y2 = 2 * np.sin(x)    # 2배 크기 파도
y3 = 3 * np.sin(x)    # 3배 크기 파도
y4 = 4 * np.sin(x)    # 4배 크기 파도

# 그래프 배경을 흰색 + 줄무늬(격자)로 꾸미기
sns.set_style('whitegrid')

# 그래프 크기 설정 (가로 10, 세로 6)
plt.figure(figsize=(10, 6))

# 네 개의 파도 그래프 한 번에 그리기
plt.plot(x, y1, x, y2, x, y3, x, y4)

# 그래프 테두리(바깥쪽 선) 숨기기 또는 보이기 설정
# 아래 설정은 오른쪽, 왼쪽 선을 보이게 함 (False = 보임)
sns.despine(right=False, left=False)  
# top=True, bottom=False는 기본값 (위쪽은 숨김, 아래쪽은 보임)

# 그래프 보여주기
plt.show()
```

```python
# seaborn을 이용해서 그래프의 글자 크기와 선 굵기 같은 '보기 스타일'을 바꾸기
# 'talk'는 발표하거나 말하면서 보여줄 때 쓰는 큰 글씨 스타일
# (paper → 가장 작음, notebook → 보통, talk → 큼, poster → 아주 큼)
sns.set_context('talk')

# x 값 만들기 (1부터 14까지 100개 숫자 만들기)
x = np.linspace(1, 14, 100)

# x에 사인(sin) 계산해서 파도 모양 만들기
y1 = np.sin(x)        # 1배 크기 파도
y2 = 2 * np.sin(x)    # 2배 크기 파도
y3 = 3 * np.sin(x)    # 3배 크기 파도
y4 = 4 * np.sin(x)    # 4배 크기 파도

# 네 개의 파도 모양 그래프 그리기
plt.plot(x, y1, x, y2, x, y3, x, y4)

# 그래프 화면에 보여주기
plt.show()
```

```python
#sns 패키지의 filghts data set 을 이용해서 heatmap 작성
flights=sns.load_dataset('flights')
flights.head()
flights.tail()
```

```python
# pivot_table을 이용해서 데이터 구조 바꾸기
# flights 데이터에서 'month'를 기준으로 행(index)으로, 'year'를 기준으로 열(columns)로 설정
# values='passengers' → 승객 수 데이터를 값으로 사용
# fill_value=0 → 데이터가 없는 곳은 0으로 채움
flight_p = pd.pivot_table(
    flights,
    index='month',       # 행(세로) 기준: 월
    columns='year',      # 열(가로) 기준: 연도
    values='passengers', # 값으로 사용할 데이터: 승객 수
    fill_value=0         # 값이 없는 곳은 0으로 채움
)

# 결과: 각 월별, 연도별 승객 수가 한눈에 보기 좋게 표 형태로 변환됨
flight_p
```

```python
# 미국 각 주별 실업률 데이터를 읽어서 단계 구분도(Choropleth)에 사용

# 1️⃣ 데이터 파일 경로
# 'folium_US_Unemployment_Oct2012.csv' → 2012년 10월 미국 주별 실업률
state_unemployment = 'C:/Users/User/Desktop/python/data/folium_US_Unemployment_Oct2012.csv'

# 2️⃣ CSV 파일 읽기
state_unemp_data = pd.read_csv(state_unemployment)

# 3️⃣ 데이터 확인 (앞쪽 5행 출력)
state_unemp_data.head()

# - state_unemp_data : 미국 50개 주별 실업률 데이터
```

```python
# folium을 이용해서 미국 주별 실업률을 색으로 표시하는 지도 만들기

# 지도 생성
# location=[위도, 경도] → 지도 중심 좌표 (미국 중앙)
# zoom_start=3 → 전국 단위로 보기 위한 확대 정도
map = folium.Map(location=[40, -98], zoom_start=3)

# Choropleth 지도 추가
# geo_data → 지도 경계 정보 (GeoJSON 형식)
# data → 실업률 데이터가 담긴 DataFrame
# columns=['State','Unemployment'] → 데이터에서 사용할 열
# key_on='feature.id' → geo_data와 data 연결 기준
# fill_color='YlGn' → 색상 스타일 (노랑-초록 계열)
# legend_name='Unemployment Rate(%)' → 범례 이름
folium.Choropleth(
    geo_data=state_geo,
    data=state_unemp_data,
    columns=['State','Unemployment'],
    key_on='feature.id',
    fill_color='YlGn',
    legend_name='Unemployment Rate(%)'
).add_to(map)

# LayerControl 추가
# 지도에 여러 레이어가 있을 때, 체크박스로 표시/해제 가능
folium.LayerControl().add_to(map)

# 지도 화면에 표시
map
```

```python
# folium과 json을 이용해서 한국 지도에 단계 구분도(Choropleth) 그리기

import json
import folium
import pandas as pd

# 1️⃣ 한국 지도 경계 데이터(json) 가져오기
# geo_path → JSON 파일 경로
# 한글이 포함된 경우, encoding='utf-8'로 읽어야 깨지지 않음
geo_path = 'C:/Users/User/Desktop/python/data/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))  # 지도 경계 데이터를 변수에 저장

# 2️⃣ 지도 생성
# 서울 중심 좌표: [위도, 경도] = [37.5502, 126.982]
# zoom_start=11 → 서울시 정도 확대
# tiles='OpenStreetMap' → 지도 스타일
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='OpenStreetMap')

# 지도를 화면에 표시
map

# 3️⃣ 단계 구분도(Choropleth)용 데이터 읽어오기
# CCTV 설치 결과 데이터 예제
# index_col='구별' → 구 이름을 행 인덱스로 사용
f_test = pd.read_csv('C:/Users/User/Desktop/python/data/cctv_result.csv', encoding='utf-8', index_col='구별')

# 데이터 확인 (앞쪽 5행)
f_test.head()

# ⚡ 설명
# - geo_str → 서울시 각 구별 경계 정보
# - f_test → 구별 CCTV 설치 현황 데이터
# - 이후 folium.Choropleth()로 연결해서 색으로 표현 가능
```

```python
# folium을 이용해서 서울시 구별 CCTV 현황 단계 구분도(Choropleth) 만들기

# Choropleth 지도 추가
# geo_data → 지도 경계 데이터 (서울시 각 구)
# data → 구별 데이터 (CCTV 설치 현황)
# columns=[f_test.index, '소계'] → index(구 이름)와 값('소계') 사용
# key_on='feature.id' → geo_data와 data 연결 기준
# fill_color='PuRd' → 색상 스타일 (보라-분홍 계열)
# legend_name='CCTV 현황' → 범례 이름
folium.Choropleth(
    geo_data=geo_str,
    data=f_test,
    columns=[f_test.index, '소계'],
    key_on='feature.id',
    fill_color='PuRd',
    legend_name='CCTV 현황'
).add_to(map)

# LayerControl 추가
# 지도에 여러 레이어가 있을 때, 체크박스로 켜고 끌 수 있음
folium.LayerControl().add_to(map)

# 지도 화면에 표시
map
```
