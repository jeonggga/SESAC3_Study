```python
# --- 1. "Tistory ìš”ì²­ ë° ì‘ë‹µ ë¶„ì„" (Tistory Request and Analysis) ----


# Tistory í™ˆí˜ì´ì§€ì˜ URLì„ 'url' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
url = 'https://www.tistory.com'

# "requests.get(url)" í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ í•´ë‹¹ URLë¡œ HTTP GET ìš”ì²­ì„ ë³´ëƒ…ë‹ˆë‹¤.
# ì„œë²„ë¡œë¶€í„° ë°›ì€ "Response ê°ì²´"ë¥¼ 'response' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
response = requests.get(url)

# Response ê°ì²´ì˜ "status_code" ì†ì„±ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
# "status_code"ëŠ” ìš”ì²­ì˜ ì„±ê³µ/ì‹¤íŒ¨ ì—¬ë¶€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” HTTP ì‘ë‹µ ì½”ë“œì…ë‹ˆë‹¤.
# "200"ì´ ì¶œë ¥ë˜ë©´ ìš”ì²­ì´ "ì„±ê³µì "ì´ì—ˆìŒì„ ì˜ë¯¸í•©ë‹ˆë‹¤.
print(response.status_code)

# Response ê°ì²´ì˜ "text" ì†ì„±(ì‘ë‹µìœ¼ë¡œ ë°›ì€ HTML ì†ŒìŠ¤ ì½”ë“œ) ì¤‘ 
# ì¸ë±ìŠ¤ 5036ë¶€í„° 6983ê¹Œì§€ì˜ ë¬¸ìì—´ ì¡°ê°ì„ ì˜ë¼ë‚´ì–´ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì´ ì½”ë“œëŠ” ì›¹í˜ì´ì§€ì˜ "íŠ¹ì • ë¶€ë¶„"ë§Œ í™•ì¸í•˜ê³ ì í•  ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.
print(response.text[5036:6983])
```

```python
# --- 2. "Google ìš”ì²­ ë° ì‘ë‹µ ë¶„ì„" (Google Request and Analysis) ---


# Google Koreaë¡œ HTTP GET ìš”ì²­ì„ ë³´ë‚´ê³ , Response ê°ì²´ë¥¼ 'r' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
r = requests.get("https://www.google.co.kr")

# Response ê°ì²´ ìì²´ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤. (ì˜ˆ: "<Response [200]>")
print(r)

# Response ê°ì²´ì˜ "text" ì†ì„±(HTML ì†ŒìŠ¤ ì½”ë“œ ì „ì²´)ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
# Googleì˜ HTMLì€ ëŒ€ë¶€ë¶„ JavaScriptë¡œ ì´ë£¨ì–´ì ¸ ìˆì–´ ì½ê¸° ì–´ë ¤ìš¸ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
print(r.text)

# Response ê°ì²´ì˜ HTML ì†ŒìŠ¤ ì½”ë“œ ë¬¸ìì—´ì˜ "ì´ ê¸¸ì´"(ë¬¸ì ìˆ˜)ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
print(len(r.text))

# HTML ì†ŒìŠ¤ ì½”ë“œì˜ "ì•ì—ì„œ 100ì"ê¹Œì§€ë§Œ ì˜ë¼ë‚´ì–´ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì´ëŠ” ì‘ë‹µì´ ì œëŒ€ë¡œ ì™”ëŠ”ì§€ ë¹ ë¥´ê²Œ í™•ì¸í•˜ëŠ” ìš©ë„ë¡œ ì‚¬ìš©ë©ë‹ˆë‹¤.
print(r.text[0:100])
```

```python
# --- 3. "Naver ìš”ì²­ ë° ì‘ë‹µ ë¶„ì„" (Naver Request and Analysis) ---


# Naver í™ˆí˜ì´ì§€ì˜ URLì„ 'url' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
url = 'http://www.naver.com'

# Naverë¡œ HTTP GET ìš”ì²­ì„ ë³´ë‚´ê³ , Response ê°ì²´ë¥¼ 'res' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
res = requests.get(url)

# Response ê°ì²´ì˜ "status_code"ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤. ("200"ì´ë©´ ì„±ê³µ)
print(res.status_code)

# Response ê°ì²´ì˜ "text" ì†ì„±(HTML ì†ŒìŠ¤ ì½”ë“œ ì „ì²´)ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
print(res.text)

# HTML ì†ŒìŠ¤ ì½”ë“œ ë¬¸ìì—´ì˜ "ì´ ê¸¸ì´"ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
print(len(res.text))
```

```python
# --- "requests" ëª¨ë“ˆì„ ì´ìš©í•œ GET ìš”ì²­ ë°©ì‹ ---



# 1. "URLì— ì§ì ‘ íŒŒë¼ë¯¸í„° í¬í•¨" ë°©ì‹ (Direct Parameter in URL)

# ë„¤ì´ë²„ ìŠ¤í¬ì¸  ë‰´ìŠ¤ì˜ ê¸°ë³¸ URLì„ 'base_url' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
base_url = "https://sports.news.naver.com/news" 

# ì¿¼ë¦¬ ìŠ¤íŠ¸ë§(Query String, íŒŒë¼ë¯¸í„°)ì´ "ì§ì ‘ í¬í•¨ëœ" ìµœì¢… URLì„ 'param_url' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
# '?' ë’¤ì˜ 'oid=079&aid=0004089508' ë¶€ë¶„ì´ ì„œë²„ë¡œ ì „ë‹¬í•  ë°ì´í„°(íŒŒë¼ë¯¸í„°)ì…ë‹ˆë‹¤.
param_url ="https://sports.news.naver.com/news?oid=079&aid=0004089508"



# "requests.get(param_url)"ì„ ì‚¬ìš©í•˜ì—¬ íŒŒë¼ë¯¸í„°ê°€ í¬í•¨ëœ ì „ì²´ URLë¡œ HTTP GET ìš”ì²­ì„ ë³´ëƒ…ë‹ˆë‹¤.
res1 = requests.get(param_url)

# Response ê°ì²´ 'res1'ì˜ "status_code"ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
# "200"ì´ ì¶œë ¥ë˜ë©´ ìš”ì²­ì´ "ì„±ê³µì "ì´ì—ˆìŒì„ ì˜ë¯¸í•˜ë©°, í˜ì´ì§€ ë°ì´í„°ë¥¼ ì •ìƒì ìœ¼ë¡œ ë°›ì•˜ë‹¤ëŠ” ëœ»ì…ë‹ˆë‹¤.
print(res1.status_code)
```

```python
# 2. "params ì¸ìˆ˜ë¥¼ ì‚¬ìš©í•œ íŒŒë¼ë¯¸í„° ì „ë‹¬" ë°©ì‹ (Using the params Argument)


# ë„¤ì´ë²„ ìŠ¤í¬ì¸  ë‰´ìŠ¤ì˜ ê¸°ë³¸ URLì…ë‹ˆë‹¤. (ì¿¼ë¦¬ ìŠ¤íŠ¸ë§ì€ í¬í•¨í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.)
# base_url = "https://sports.news.naver.com/news"

# ì„œë²„ë¡œ ì „ë‹¬í•  íŒŒë¼ë¯¸í„°ë¥¼ Python "ë”•ì…”ë„ˆë¦¬" í˜•íƒœë¡œ ì •ì˜í•©ë‹ˆë‹¤.
# key: value ìŒì´ ì›¹ì˜ íŒŒë¼ë¯¸í„° ì´ë¦„ê³¼ ê°’ìœ¼ë¡œ ë§¤í•‘ë©ë‹ˆë‹¤.
param = {'oid':'079', 'aid':'0004089508'}

# "requests.get(base_url, params=param)" í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ GET ìš”ì²­ì„ ë³´ëƒ…ë‹ˆë‹¤.
# "requests" ëª¨ë“ˆì´ "base_url" ë’¤ì— ë”•ì…”ë„ˆë¦¬ 'param'ì˜ ë‚´ìš©ì„ ìë™ìœ¼ë¡œ ë¶™ì—¬ ì¿¼ë¦¬ ìŠ¤íŠ¸ë§ì„ ì™„ì„±í•©ë‹ˆë‹¤.
# ì¦‰, requestsê°€ ë‚´ë¶€ì ìœ¼ë¡œ 'base_url?oid=079&aid=0004089508' í˜•íƒœì˜ URLì„ ë§Œë“¤ì–´ ìš”ì²­í•©ë‹ˆë‹¤.
# ì´ ë°©ì‹ì€ í•œê¸€ì´ë‚˜ íŠ¹ìˆ˜ë¬¸ìê°€ í¬í•¨ëœ íŒŒë¼ë¯¸í„° ì²˜ë¦¬(URL ì¸ì½”ë”©) ì‹œ "ë§¤ìš° ì•ˆì „"í•˜ê³  "ê°€ë…ì„±"ì´ ì¢‹ìŠµë‹ˆë‹¤.
res2 = requests.get(base_url, params=param)

# Response ê°ì²´ 'res2'ì˜ "status_code"ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤. (ì—­ì‹œ "200"ì´ ì˜ˆìƒë©ë‹ˆë‹¤.)
print(res2.status_code)
```

```python
# --- URL ì—°ê²° ë° ë°ì´í„° ì½ê¸° ---


# 1. "URL ì—°ê²° ë° ì‘ë‹µ ê°ì²´ í™•ì¸" (URL Connection and Response Object)

# ë„¤ì´ë²„ í™ˆí˜ì´ì§€ì˜ URLì„ 'url' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
url ='http://www.naver.com'

# "urlopen(url)" í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ í•´ë‹¹ URLë¡œ ì—°ê²°ì„ ì‹œë„í•˜ê³ ,
# ì„œë²„ë¡œë¶€í„° ë°›ì€ "HTTPResponse ê°ì²´"ë¥¼ 'html' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
# ì´ ê°ì²´ëŠ” ì›¹ ì„œë²„ì™€ì˜ ì—°ê²°ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
# âš ï¸ ì£¼ì˜: ì´ ì½”ë“œëŠ” 'urlopen'ì´ ì„í¬íŠ¸ë˜ì—ˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.
html = urlopen(url)

# "HTTPResponse ê°ì²´"ì¸ 'html' ë³€ìˆ˜ì˜ ì •ë³´ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì¶œë ¥ ê²°ê³¼ëŠ” ë³´í†µ <http.client.HTTPResponse object at 0x...> í˜•íƒœë¡œ ë‚˜íƒ€ë‚˜ë©°,
# ì´ëŠ” ì„œë²„ì™€ì˜ ì—°ê²°ì´ ì„±ê³µí–ˆìŒì„ ì˜ë¯¸í•©ë‹ˆë‹¤.
print(html)
```

```python
# 2. "ë°”ì´ë„ˆë¦¬ ë°ì´í„° ì½ê¸° ë° íƒ€ì… í™•ì¸" (Reading Binary Data and Type Check)


# 'html' ê°ì²´(HTTPResponse)ì˜ ".read()" ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ 
# ì„œë²„ê°€ ë³´ë‚¸ ì‘ë‹µ ë³¸ë¬¸(HTML ì†ŒìŠ¤ ì½”ë“œ) "ì „ì²´"ë¥¼ ì½ì–´ 'rawtext'ì— ì €ì¥í•©ë‹ˆë‹¤.
# ì›¹ì—ì„œ ì½ì–´ì˜¨ ë°ì´í„°ëŠ” ê¸°ë³¸ì ìœ¼ë¡œ "ë°”ì´ë„ˆë¦¬(bytes)" í˜•íƒœë¡œ ì €ì¥ë©ë‹ˆë‹¤.
rawtext = html.read()

# 'rawtext' ë³€ìˆ˜ì˜ "íƒ€ì…"ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì¶œë ¥ ê²°ê³¼ëŠ” "<class 'bytes'>"ì´ë©°, ì´ëŠ” ë°ì´í„°ê°€ 0ê³¼ 1ë¡œ ì´ë£¨ì–´ì§„ "ë°”ì´íŠ¸ ì‹œí€€ìŠ¤"ì„ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
print(type(rawtext))

# 'rawtext' ë³€ìˆ˜ì˜ "íƒ€ì…"ì„ ì¶œë ¥í•˜ê³  ì¤„ ë°”ê¿ˆ ë¬¸ìë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤. (ì¬í™•ì¸)
print(type(rawtext),"\n")

# 'rawtext' ë³€ìˆ˜ë¥¼ "bytes" íƒ€ì…ìœ¼ë¡œ ë³€í™˜(ì´ë¯¸ bytes íƒ€ì…ì´ë¯€ë¡œ í˜•íƒœ ìœ ì§€)í•˜ê³ , 
# ë°ì´í„°ì˜ "ì•ì—ì„œ 200 ë°”ì´íŠ¸"ê¹Œì§€ë§Œ ì˜ë¼ë‚´ì–´ ì¶œë ¥í•©ë‹ˆë‹¤.
# ë°”ì´ë„ˆë¦¬ ë°ì´í„°ëŠ” ì‚¬ëŒì´ ì½ì„ ìˆ˜ ì—†ëŠ” í˜•íƒœë¡œ ë³´ì…ë‹ˆë‹¤ (ì˜ˆ: b'<!DOCTYPE html>...').
print(bytes(rawtext)[:200], "\n")

# 'rawtext' (ë°”ì´ë„ˆë¦¬ ë°ì´í„°)ë¥¼ ".decode('utf8')" ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ 
# 'utf8' ì¸ì½”ë”© ë°©ì‹ì— ë”°ë¼ "ë¬¸ìì—´(string)"ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
# ë³€í™˜ëœ ë¬¸ìì—´ì˜ "ì•ì—ì„œ 200ì"ê¹Œì§€ë§Œ ì˜ë¼ë‚´ì–´ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì´ì œ ë°ì´í„°ê°€ ì‚¬ëŒì´ ì½ì„ ìˆ˜ ìˆëŠ” HTML í…ìŠ¤íŠ¸ í˜•íƒœë¡œ ì¶œë ¥ë©ë‹ˆë‹¤.
print(rawtext.decode('utf8')[:200])


# "í•µì‹¬ ê°œë…: ë°”ì´íŠ¸ vs ë¬¸ìì—´"
#  - ì›¹ì—ì„œ ë°ì´í„°ë¥¼ ì½ì–´ì˜¤ë©´ í•­ìƒ "ë°”ì´íŠ¸(bytes)" í˜•íƒœë¡œ ë°›ê²Œ ë©ë‹ˆë‹¤.
#  - **"ë°”ì´íŠ¸"**ëŠ” ì»´í“¨í„°ê°€ ì´í•´í•˜ëŠ” 0ê³¼ 1ì˜ ì‹œí€€ìŠ¤ì´ë©°, íŒŒì¼ ì €ì¥ì´ë‚˜ ë„¤íŠ¸ì›Œí¬ ì „ì†¡ì— ì‚¬ìš©ë©ë‹ˆë‹¤.
#  - ìš°ë¦¬ê°€ ì½ì„ ìˆ˜ ìˆëŠ” í…ìŠ¤íŠ¸ë¡œ ì‚¬ìš©í•˜ë ¤ë©´ .decode() ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬
# íŠ¹ì • ì¸ì½”ë”©(ì—¬ê¸°ì„œëŠ” utf8)ì„ í†µí•´ **"ë¬¸ìì—´(str)"**ë¡œ ë³€í™˜í•´ì•¼ í•©ë‹ˆë‹¤.
```

```python
# --- URL ì—°ê²°, ë°ì´í„° ì½ê¸° ë° BeautifulSoup íŒŒì‹± ---



# 1. "URL ì—°ê²° ë° ì‘ë‹µ ê°ì²´ í™•ì¸" (URL Connection and Response Object)

# ë„¤ì´ë²„ í™ˆí˜ì´ì§€ì˜ URLì„ 'url' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
url = 'http://www.naver.com'

# "urlopen(url)" í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ í•´ë‹¹ URLë¡œ ì—°ê²°ì„ ì‹œë„í•˜ê³ ,
# ì„œë²„ë¡œë¶€í„° ë°›ì€ "HTTPResponse ê°ì²´"ë¥¼ 'html' ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
# ì´ ê°ì²´ëŠ” ì›¹ ì„œë²„ì™€ì˜ "í™œì„± ì—°ê²°"ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
# ì£¼ì˜: 'urlopen'ì€ urllib.request ëª¨ë“ˆì— ì†í•©ë‹ˆë‹¤.
html = urlopen(url)

# "HTTPResponse ê°ì²´"ì¸ 'html' ë³€ìˆ˜ì˜ ì •ë³´ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì¶œë ¥ ê²°ê³¼ëŠ” <http.client.HTTPResponse object at 0x...> í˜•íƒœì´ë©°, 
# ì´ëŠ” ì„œë²„ì™€ì˜ ì—°ê²°ì´ "ì„±ê³µì "ì´ì—ˆìŒì„ ì˜ë¯¸í•©ë‹ˆë‹¤.
print(html)
```

```python
# 2. "BeautifulSoup ê°ì²´ ìƒì„± ë° íƒ€ì… í™•ì¸" (BeautifulSoup Object Creation and Type Check)


# "bs4.BeautifulSoup()" ìƒì„±ìë¥¼ ì‚¬ìš©í•˜ì—¬ HTML íŒŒì‹± ê°ì²´ë¥¼ ë§Œë“­ë‹ˆë‹¤.
# ì²« ë²ˆì§¸ ì¸ìˆ˜ì¸ 'html' (HTTPResponse ê°ì²´)ì—ì„œ HTML ë°ì´í„°ë¥¼ ì½ì–´ì˜µë‹ˆë‹¤.
# ë‘ ë²ˆì§¸ ì¸ìˆ˜ì¸ 'html.parser'ëŠ” íŒŒì´ì¬ í‘œì¤€ ë¼ì´ë¸ŒëŸ¬ë¦¬ì— í¬í•¨ëœ "HTML íŒŒì„œ"ë¥¼ ì‚¬ìš©í•˜ë„ë¡ ì§€ì •í•©ë‹ˆë‹¤.
# íŒŒì‹±ëœ êµ¬ì¡°í™”ëœ ê°ì²´ëŠ” 'bs_obj' ë³€ìˆ˜ì— ì €ì¥ë©ë‹ˆë‹¤.
bs_obj = bs4.BeautifulSoup(html,'html.parser')

# 'bs_obj' ë³€ìˆ˜ì˜ "íƒ€ì…"ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì¶œë ¥ ê²°ê³¼ëŠ” "<class 'bs4.BeautifulSoup'>"ì´ë©°, ì´ëŠ” ë°ì´í„°ê°€ 
# ì›¹ ìŠ¤í¬ë˜í•‘ì— ìš©ì´í•œ "BeautifulSoup ê°ì²´"ë¡œ ë³€í™˜ë˜ì—ˆìŒì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
print(type(bs_obj))
```

```python
# 3. "ë°ì´í„° ê²€ìƒ‰ ë° í…ìŠ¤íŠ¸ ì¶”ì¶œ" (Data Search and Text Extraction)


# 'bs_obj' ê°ì²´ì˜ ".find('div')" ë©”ì„œë“œë¥¼ í˜¸ì¶œí•©ë‹ˆë‹¤.
# ì´ëŠ” BeautifulSoup íŠ¸ë¦¬ êµ¬ì¡°ì—ì„œ "ì²« ë²ˆì§¸ë¡œ ë°œê²¬ë˜ëŠ” 'div' íƒœê·¸ ìš”ì†Œ"ë¥¼ ì°¾ì•„ ë°˜í™˜í•©ë‹ˆë‹¤.
# ì¶œë ¥ ê²°ê³¼ëŠ” <div>hello</div> í˜•íƒœì˜ íƒœê·¸ ê°ì²´ì…ë‹ˆë‹¤.
print(bs_obj.find('div'))

# ".find('div')"ë¡œ ì°¾ì€ 'div' íƒœê·¸ ê°ì²´ ë’¤ì— ".text" ì†ì„±ì„ ë¶™ì…ë‹ˆë‹¤.
# ".text" ì†ì„±ì€ í•´ë‹¹ íƒœê·¸ ê°ì²´ê°€ í¬í•¨í•˜ê³  ìˆëŠ” "ìˆœìˆ˜í•œ í…ìŠ¤íŠ¸ ë‚´ìš©"ë§Œì„ ì¶”ì¶œí•©ë‹ˆë‹¤.
# ì¶œë ¥ ê²°ê³¼ëŠ” 'hello' ë¬¸ìì—´ì…ë‹ˆë‹¤.
print(bs_obj.find('div').text)
```

```python
# --- BeautifulSoupì„ ì´ìš©í•œ ë°ì´í„° ëª©ë¡(List) ì¶”ì¶œ ---


# 1. "HTML ë¬¸ìì—´ ì •ì˜ ë° ê°ì²´ ìƒì„±" (HTML Definition and Object Creation)

# ì—¬ëŸ¬ ê°œì˜ ëª©ë¡ í•­ëª©(<li>)ì„ í¬í•¨í•˜ëŠ” HTML ë¬¸ìì—´ì„ ì •ì˜í•©ë‹ˆë‹¤.
html_str = """
<!DOCTYPE html>
<html>
    <body>
        <ul>
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
    </body>
</html>
"""

# BeautifulSoup ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
# 'html_str' ë¬¸ìì—´ì„ 'html.parser'ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¶„ì„í•˜ê³ , ê·¸ ê²°ê³¼ë¥¼ 'bs_obj'ì— ì €ì¥í•©ë‹ˆë‹¤.
bs_obj = bs4.BeautifulSoup(html_str,'html.parser')
bs_obj
```

```python
# 2. "ìµœìƒìœ„ ëª©ë¡ íƒœê·¸ ê²€ìƒ‰" (Finding the Top-Level List Tag)


# "bs_obj.find('ul')" ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ë¬¸ì„œì—ì„œ "ì²« ë²ˆì§¸ë¡œ ë°œê²¬ë˜ëŠ” 'ul' íƒœê·¸ ìš”ì†Œ" ì „ì²´ë¥¼ ì°¾ìŠµë‹ˆë‹¤.
# 'ul_t' ë³€ìˆ˜ëŠ” <ul>...</ul> ì „ì²´ë¥¼ í¬í•¨í•˜ëŠ” BeautifulSoup íƒœê·¸ ê°ì²´ê°€ ë©ë‹ˆë‹¤.
ul_t = bs_obj.find('ul')
# ì°¾ì€ 'ul' íƒœê·¸ ê°ì²´ì˜ ì „ì²´ ë‚´ìš©ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
print(ul_t)

# 'ul_t' ê°ì²´ ë‚´ì—ì„œ ".find('li')" ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ "ì²« ë²ˆì§¸ë¡œ ë°œê²¬ë˜ëŠ” 'li' íƒœê·¸ ìš”ì†Œ"ë¥¼ ì°¾ìŠµë‹ˆë‹¤.
# ì¶œë ¥ ì—†ì´ ì‹¤í–‰ë˜ì—ˆì§€ë§Œ, ê²°ê³¼ëŠ” '<li>hello</li>' ê°ì²´ê°€ ë©ë‹ˆë‹¤.
ul_t.find('li')
```

```python
# 3. "ëª¨ë“  ëª©ë¡ í•­ëª© ì¶”ì¶œ" (Extracting All List Items)


# 'ul_t' ê°ì²´ ë‚´ì—ì„œ ".findAll('li')" ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ "ëª¨ë“  'li' íƒœê·¸ ìš”ì†Œ"ë¥¼ ê²€ìƒ‰í•©ë‹ˆë‹¤.
# ì°¾ì€ ëª¨ë“  ìš”ì†ŒëŠ” ë¦¬ìŠ¤íŠ¸ í˜•íƒœë¡œ 'uls' ë³€ìˆ˜ì— ì €ì¥ë©ë‹ˆë‹¤.
uls = ul_t.findAll('li')
# 'uls' ë¦¬ìŠ¤íŠ¸ì˜ ë‚´ìš©ì„ ì¶œë ¥í•©ë‹ˆë‹¤. (ê° <li> íƒœê·¸ ê°ì²´ë“¤ì´ ë¦¬ìŠ¤íŠ¸ì˜ ìš”ì†Œë¡œ í¬í•¨ë¨)
print(uls)


# 'uls' ë¦¬ìŠ¤íŠ¸ì˜ "ì„¸ ë²ˆì§¸ ìš”ì†Œ" (ì¸ë±ìŠ¤ 2)ì˜ "íƒ€ì…"ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
# ì¶œë ¥ ê²°ê³¼ëŠ” "<class 'bs4.element.Tag'>"ë¡œ, ë¦¬ìŠ¤íŠ¸ ìš”ì†Œê°€ íƒœê·¸ ê°ì²´ì„ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.
print(type(uls[2]))
# ì„¸ ë²ˆì§¸ 'li' íƒœê·¸ ê°ì²´ì—ì„œ ".text" ì†ì„±ì„ ì‚¬ìš©í•˜ì—¬ "ìˆœìˆ˜í•œ í…ìŠ¤íŠ¸"('welcome')ë§Œ ì¶”ì¶œí•˜ì—¬ ì¶œë ¥í•©ë‹ˆë‹¤.
print(uls[2].text)


# 'uls' ë¦¬ìŠ¤íŠ¸ì˜ ê° 'li' íƒœê·¸ ê°ì²´ì— ëŒ€í•´ ë°˜ë³µë¬¸(for loop)ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.
for li in uls :
    # ê° 'li' ê°ì²´ì—ì„œ ".text"ë¥¼ ì‚¬ìš©í•˜ì—¬ í…ìŠ¤íŠ¸ ë‚´ìš©ì„ ì¶”ì¶œí•˜ê³  ì¶œë ¥í•©ë‹ˆë‹¤.
    # ì´ ê³¼ì •ìœ¼ë¡œ 'hello', 'bye', 'welcome'ì´ ê°ê° í•œ ì¤„ì”© ì¶œë ¥ë©ë‹ˆë‹¤.
    print(li.text)
```

```python
# --- BeautifulSoupì„ ì´ìš©í•œ í•˜ì´í¼ë§í¬ ë°ì´í„° ì¶”ì¶œ ë° DataFrame ìƒì„± ---


# 1. "HTML ì •ì˜ ë° BeautifulSoup ê°ì²´ ìƒì„±" (HTML Definition and Parsing)

# í´ë˜ìŠ¤ê°€ 'ko'ì™€ 'sns'ì¸ ë‘ ê°œì˜ <ul> ëª©ë¡ì— ì—¬ëŸ¬ ê°œì˜ í•˜ì´í¼ë§í¬(<a> íƒœê·¸)ë¥¼ í¬í•¨í•˜ëŠ” HTML ë¬¸ìì—´ì„ ì •ì˜í•©ë‹ˆë‹¤.
html_str = """
<!DOCTYPE html>
<html>
    <body>
        <ul class="ko">
            <li><a href="https://www.naver.com/">ë„¤ì´ë²„</a></li>
            <li><a href="https://www.daum.net/">ë‹¤ìŒ</a></li>
        </ul>
        <ul class="sns">
            <li><a href="https://www.goole.com/">êµ¬ê¸€</a></li>
            <li><a href="https://www.facebook.net/">í˜ì´ìŠ¤ë¶</a></li>
        </ul>
    </body>
</html>
"""

# BeautifulSoup ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. 'html_str'ì„ 'html.parser'ë¡œ ë¶„ì„í•˜ê³  'bs_obj'ì— ì €ì¥í•©ë‹ˆë‹¤.
bs_obj = bs4.BeautifulSoup(html_str,'html.parser')
bs_obj
```

```python
# 3. "ë°˜ë³µë¬¸ì„ ì´ìš©í•œ URL ì¶”ì¶œ" (Extracting URLs via Loop)


hrefs=[] # ì¶”ì¶œí•œ URLì„ ì €ì¥í•  ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
# bs_obj.find_all('a')ë¥¼ ì‚¬ìš©í•˜ì—¬ ì°¾ì€ ëª¨ë“  <a> íƒœê·¸ ìš”ì†Œì— ëŒ€í•´ ë°˜ë³µí•©ë‹ˆë‹¤.
for a in bs_obj.find_all('a') :
    # í˜„ì¬ <a> íƒœê·¸ì˜ 'href' ì†ì„± ê°’ì„ ì¶”ì¶œí•˜ì—¬ ì¶œë ¥í•©ë‹ˆë‹¤.
    print(a['href'])
    # ì¶”ì¶œí•œ URLì„ 'hrefs' ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€í•©ë‹ˆë‹¤.
    hrefs.append(a['href'])
```

```python
# ë„¤ì´ë²„ ë‰´ìŠ¤ ì„¹ì…˜ ë©”ë‰´ì˜ íƒœê·¸ì™€ í´ë˜ìŠ¤ í™•ì¸ (ê°œë°œì ë„êµ¬ í™œìš©)
# body > section > header > div.Nlnb._float_lnb > div > div > div .Nlnb_left ._lnb_scroll > div > div > ul
# ê°œë°œì ë„êµ¬ì—ì„œ í™•ì¸í•œ ì…€ë ‰í„°ëŠ” ë„ˆë¬´ ê¸¸ì–´ì„œ ì‚¬ìš©í•˜ê¸° ë¶ˆí¸í•¨
# ëŒ€ì‹  'ul' íƒœê·¸ì˜ 'class' ì†ì„±ì´ 'Nlnb_menu_list'ì¸ ìš”ì†Œë¥¼ ì°¾ì•„ì„œ ì‚¬ìš©

# 'ul' íƒœê·¸ ì¤‘ì—ì„œ 'class' ì†ì„±ì´ 'Nlnb_menu_list'ì¸ ëª¨ë“  ìš”ì†Œë¥¼ ì°¾ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
uls = bs_obj.findAll("ul", {"class":"Nlnb_menu_list"})

# 'uls' ë¦¬ìŠ¤íŠ¸ì˜ ê¸¸ì´ í™•ì¸ (ê²°ê³¼ì ìœ¼ë¡œ ì›ì†Œê°€ 1ê°œì¼ ê²ƒìœ¼ë¡œ ì˜ˆìƒ)
len(uls)
```

```python
# 'ul' íƒœê·¸ ì¤‘ì—ì„œ 'class' ì†ì„±ì´ 'Nlnb_menu_list'ì¸ ì²« ë²ˆì§¸ ìš”ì†Œë¥¼ ì°¾ìŒ
ul = bs_obj.find("ul", {"class":"Nlnb_menu_list"})

# ì°¾ì€ ì²« ë²ˆì§¸ 'ul' ìš”ì†Œë¥¼ ì¶œë ¥ (HTML êµ¬ì¡°ë¥¼ í™•ì¸í•  ìˆ˜ ìˆìŒ)
ul
```

```python
# 'ul' ìš”ì†Œì—ì„œ ëª¨ë“  'li' íƒœê·¸ë¥¼ ì°¾ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
# ì´ ë¦¬ìŠ¤íŠ¸ëŠ” ë©”ë‰´ í•­ëª©ì„ ë‚˜íƒ€ë‚´ëŠ” 'li' íƒœê·¸ë“¤ì´ í¬í•¨ë¨
lis = ul.findAll('li')

# 'li' ìš”ì†Œì˜ ê°œìˆ˜ë¥¼ ì¶œë ¥ (ë©”ë‰´ í•­ëª©ì˜ ê°œìˆ˜ í™•ì¸)
len(lis)
```

```python
# ê° 'li' ìš”ì†Œì— ëŒ€í•´ ë°˜ë³µë¬¸ì„ ì‹¤í–‰
for li in lis :
    
    # ê° 'li' ìš”ì†Œ ë‚´ì—ì„œ 'a' íƒœê·¸ë¥¼ ì°¾ìŒ (ë©”ë‰´ í•­ëª©ì˜ ë§í¬)
    a_tag = li.find('a')
    
    # 'a' íƒœê·¸ì—ì„œ í…ìŠ¤íŠ¸ (ë©”ë‰´ í•­ëª©ì˜ ì œëª©)ì™€ 'href' ì†ì„± (ë§í¬ URL)ì„ ì¶œë ¥
    print(a_tag.text, " : ", a_tag['href'])
```

```python
# 'li' íƒœê·¸ ì¤‘ì—ì„œ class ì†ì„±ì´ 'Nlist_item'ì¸ ëª¨ë“  ìš”ì†Œë¥¼ ì°¾ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
lis = bs_obj.findAll("li", {"class" : "Nlist_item"})

# 'lis' ë¦¬ìŠ¤íŠ¸ì˜ ê¸¸ì´ í™•ì¸ (ì¶”ì¶œëœ 'li' íƒœê·¸ì˜ ê°œìˆ˜ í™•ì¸)
len(lis)
```

```python
# 'lis' ë¦¬ìŠ¤íŠ¸ì˜ ê° 'li' ìš”ì†Œì— ëŒ€í•´ ë°˜ë³µë¬¸ì„ ì‹¤í–‰
for li in lis :
    
    # ê° 'li' ìš”ì†Œ ë‚´ì—ì„œ 'a' íƒœê·¸ë¥¼ ì°¾ì•„ì„œ 'a_tag'ì— ì €ì¥
    a_tag = li.find('a')
    
    # 'a_tag.text'ëŠ” 'a' íƒœê·¸ ì•ˆì˜ í…ìŠ¤íŠ¸(ë§í¬ ì œëª©)ë¥¼ ì¶œë ¥
    print(a_tag.text)
    
    # 'a_tag['href']'ëŠ” 'a' íƒœê·¸ì˜ 'href' ì†ì„±(ë§í¬ URL)ì„ ì¶œë ¥
    print(a_tag['href'])
```

```python
# ë©”ë‰´ í•­ëª©ì˜ í…ìŠ¤íŠ¸ì™€ ë§í¬ë¥¼ ì €ì¥í•  ë¹ˆ ë¦¬ìŠ¤íŠ¸ ì´ˆê¸°í™”
section = []
link = []

# 'lis' ë¦¬ìŠ¤íŠ¸ì˜ ê° 'li' ìš”ì†Œì— ëŒ€í•´ ë°˜ë³µë¬¸ì„ ì‹¤í–‰
for li in lis :
    
    # ê° 'li' ìš”ì†Œ ë‚´ì—ì„œ 'a' íƒœê·¸ë¥¼ ì°¾ì•„ì„œ 'a_tag'ì— ì €ì¥
    a_tag = li.find('a')
    
    # 'a_tag.text'ëŠ” ë§í¬ì˜ í…ìŠ¤íŠ¸(ë©”ë‰´ ì œëª©)ë¥¼ 'section' ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    section.append(a_tag.text)
    
    # 'a_tag['href']'ëŠ” ë§í¬ì˜ 'href' ì†ì„±(ë§í¬ URL)ì„ 'link' ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    link.append(a_tag['href'])
```

```python
# 'section'ê³¼ 'link' ë¦¬ìŠ¤íŠ¸ë¥¼ ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ê²°í•©
col_dict = {'Section' : section, "Link" : link}

# ë”•ì…”ë„ˆë¦¬ë¥¼ ì‚¬ìš©í•´ DataFrame ìƒì„±
# 'Section'ê³¼ 'Link' ì—´ì„ ê°€ì§„ pandas DataFrameì´ ìƒì„±ë¨
news_section_df = pd.DataFrame(col_dict)

# ìƒì„±ëœ DataFrame ì¶œë ¥
news_section_df
```
