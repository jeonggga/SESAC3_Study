```python
# 환경 변수 로드 및 라이브러리 임포트
import os
from dotenv import load_dotenv
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

# OpenAI 클라이언트 초기화
from openai import OpenAI
client = OpenAI()
```

```python
# 채팅 완료 요청 생성 (스트리밍 활성화)
response = client.chat.completions.create(
 model="gpt-4.1-mini", # 사용할 모델
 messages=[
 {"role": "system", "content": "You are a helpful assistant."}, # 시스템 역할 메시지
 {"role": "user", "content": "Hello! I'm John."} # 사용자 입력 메시지
 ],
 stream=True # 스트리밍 응답 사용
)

# 응답 청크를 순회하며 출력
for chunk in response:
 choice = chunk.choices[0]
 # 응답이 완료되지 않은 경우에만 내용 출력 (finish_reason이 None인 경우)
 if choice.finish_reason is None:
    print(choice.delta.content)
```
